# SSP Scenarios

Using data from the [SSP Database](https://tntcat.iiasa.ac.at/SspDb/dsd?Action=htmlpage&page=10), we can generate plots on the carbon emissions, carbon concentration, and mean global temperature increase in each of the five scenarios.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ssp_naming = {"SSP1-19":"ssp119", "SSP1-26":"ssp126", "SSP2-45":"ssp245", "SSP3-Baseline":"ssp370", "SSP5-Baseline":"ssp585"} 
co2_conc = pd.read_csv("../data/ssp/co2_concentrations.csv").drop(["Model","Region","Variable","Unit"], axis=1).set_index("Scenario").rename(ssp_naming).transpose()[ssp_naming.values()]
co2_em = pd.read_csv("../data/ssp/co2_emissions.csv").drop(["Model","Region","Variable","Unit"], axis=1).set_index("Scenario").rename(ssp_naming).transpose()[ssp_naming.values()]
temp_C = pd.read_csv("../data/ssp/temp_C.csv").drop(["Model","Region","Variable","Unit"], axis=1).set_index("Scenario").rename(ssp_naming).transpose()[ssp_naming.values()]
```

```python
fig, ax = plt.subplots(figsize=(7,5))
co2_conc.plot(ax=ax)
plt.ylabel("CO2 Concentration [ppm]")
plt.xlabel("Year")
plt.title("CO2 Concentrations in Different SSP Scenarios")
plt.grid(alpha=0.3)
```

```python
fig, ax = plt.subplots(figsize=(7,5))
co2_em.plot(ax=ax)
plt.ylabel("CO2 Emissions [Mt/year]")
plt.xlabel("Year")
plt.title("CO2 Emissions in Different SSP Scenarios")
plt.grid(alpha=0.3)
```

```python
fig, ax = plt.subplots(figsize=(7,5))
temp_C.plot(ax=ax)
plt.ylabel("Global Mean Temperature Increase [C]")
plt.xlabel("Year")
plt.title("Global Mean Temperature Increase in Different SSP Scenarios")
plt.grid(alpha=0.3)
```
